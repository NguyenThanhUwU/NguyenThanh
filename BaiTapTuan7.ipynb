W7A16: Kiểm tra và tạo mảng đẹp
Mô tả:
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. In ra y.

```python
def min_removals_to_beautiful(arr):
    from collections import Counter
    counts = Counter(arr)
    removals = 0
    for k, v in counts.items():
        if v >= k:
            removals += (v - k)
        else:
            removals += v
    return removals

arr = list(map(int, input().split()))
print(min_removals_to_beautiful(arr))
```

W7A17: Thuật toán Boyer-Moore
Mô tả:
Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1) - Boyer–Moore. Nếu không có phần tử nào thỏa mãn thì không in gì.

```python
def boyer_moore(arr):
    candidate = None
    count = 0
    for num in arr:
        if count == 0:
            candidate = num
        count += 1 if num == candidate else -1
    
    if arr.count(candidate) > len(arr) // 2:
        return candidate
    return ""

arr = list(map(int, input().split()))
res = boyer_moore(arr)
if res != "":
    print(res)
```

W7A19: Merge 2 mảng đã sắp xếp
Mô tả:
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào.

```python
def get_asc(arr):
    if len(arr) > 1 and arr[0] > arr[-1]:
        return arr[::-1]
    return arr

def merge_sorted(arr1, arr2):
    arr1 = get_asc(arr1)
    arr2 = get_asc(arr2)
    res = []
    i = 0
    j = 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] <= arr2[j]:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    return res

arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
print(*(merge_sorted(arr1, arr2)))
```

W7A21: Tính ngày của năm 2020

```python
def day_of_year_2020(d, m):
    days_in_month = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    if 1 <= m <= 12 and 1 <= d <= days_in_month[m]:
        return sum(days_in_month[:m]) + d
    return "Invalid"

n = int(input())
t = int(input())
res = day_of_year_2020(n, t)
print(res)
```

W7A22: Tìm tên sinh viên hay gặp nhất
Mô tả:
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
from collections import Counter

n = int(input())
names = []
for _ in range(n):
    full_name = input().strip()
    if full_name:
        names.append(full_name.split()[-1])

counts = Counter(names)
if counts:
    max_freq = max(counts.values())
    candidates = [name for name, freq in counts.items() if freq == max_freq]
    print(min(candidates))
```

W7A25: Tìm số lớn nhất có k chữ số
Mô tả:
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số. Dòng 1 là xâu số, dòng 2 là k.

```python
s = input().strip()
k = int(input())
to_drop = len(s) - k
stack = []

for digit in s:
    while to_drop > 0 and stack and stack[-1] < digit:
        stack.pop()
        to_drop -= 1
    stack.append(digit)

print("".join(stack[:k]))
```

W7A27: Dãy ngoặc hợp lệ
Mô tả:
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không. In ra True hoặc False.

```python
s = input().strip()
stack = []
mapping = {")": "(", "}": "{", "]": "["}
is_valid = True

for char in s:
    if char in mapping.values():
        stack.append(char)
    elif char in mapping:
        if not stack or stack[-1] != mapping[char]:
            is_valid = False
            break
        stack.pop()

if stack:
    is_valid = False
print(is_valid)
```

W7A28: Dãy hợp số liên tiếp dài nhất
Mô tả:
Tìm dãy các hợp số dài nhất liên tiếp trong mảng không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua). Dòng 1 nhập mảng, dòng 2 nhập N

```python
def is_composite(num):
    if num <= 3:
        return False
    if num % 2 == 0 or num % 3 == 0:
        return True
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return True
        i += 6
    return False

arr = list(map(int, input().split()))
n = int(input())

max_seq = []
curr_seq = []

for num in arr:
    if num > n and is_composite(num):
        curr_seq.append(num)
    else:
        if len(curr_seq) > len(max_seq):
            max_seq = curr_seq
        curr_seq = []

if len(curr_seq) > len(max_seq):
    max_seq = curr_seq

if max_seq:
    print(*max_seq)
```

W7A29: Dãy con tăng dần liên tục dài nhất
Mô tả:
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau.

```python
arr = list(map(int, input().split()))
if not arr:
    print()
else:
    max_seq = []
    curr_seq = [arr[0]]
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            curr_seq.append(arr[i])
        else:
            if len(curr_seq) > len(max_seq):
                max_seq = curr_seq
            curr_seq = [arr[i]]
    if len(curr_seq) > len(max_seq):
        max_seq = curr_seq
    print(*max_seq)
```

W7A30: Dãy con tăng dần KHÔNG liên tục dài nhất
Mô tả:
Tìm Dãy con tăng dần KHÔNG liên tục dài nhất (LIS) của một danh sách nhập vào từ bàn phím.

```python
arr = list(map(int, input().split()))
if not arr:
    print()
else:
    n = len(arr)
    dp = [1] * n
    parent = [-1] * n
    
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                parent[i] = j
    
    max_len = max(dp)
    max_idx = dp.index(max_len)
    
    lis = []
    while max_idx != -1:
        lis.append(arr[max_idx])
        max_idx = parent[max_idx]
    
    print(*lis[::-1])
```
