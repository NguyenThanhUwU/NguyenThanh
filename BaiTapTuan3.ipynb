W3A1 
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý. 


```python
n = int(input())

if 100 <= n <= 999:
    a = n // 100
    b = (n % 100) // 10
    c = n % 10
    dao_nguoc = c * 100 + b * 10 + a
    print(dao_nguoc)
```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 
hay không. 
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n = int(input())

if n > 0 and (n & (n - 1)) == 0:
    print("Yes")
else:
    print("No")
```

tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in

```python
a = int(input())
b = int(input())
c = int(input())

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    else:
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
```

W3A16 
[Rounding] 
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số 
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả 
cách nhau một dấu cách

```python
x = float(input())

if x >= 0:
    tron_xuong = int(x)
    if x == tron_xuong:
        tron_len = tron_xuong
    else:
        tron_len = tron_xuong + 1
    gan_nhat = int(x + 0.5)
else:
    tron_len = int(x)
    if x == tron_len:
        tron_xuong = tron_len
    else:
        tron_xuong = tron_len - 1
    gan_nhat = int(x - 0.5)

print(f"{tron_len} {tron_xuong} {gan_nhat}")
```

W3A17 
Cấp số nhân  
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 
cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?  
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội 
và so sánh vs c, c nhân công bội và so sánh vs d. 

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())

if a != 0 and b % a == 0:
    q = b // a
    if b * q == c and c * q == d:
        print("Yes")
    else:
        print("No")
elif a == 0 and b == 0 and c == 0 and d == 0:
    print("Yes")
else:
    print("No")
```
