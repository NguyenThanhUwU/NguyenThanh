Nhập vào một số nguyên đại diện cho tháng trong năm
in ra số ngày trong tháng đó

```python
thang = int(input("Nhập tháng: "))
if thang == 1 or thang == 3 or thang == 5 or thang == 7 or thang == 8 or thang == 10 or thang == 12:
    s = f"Thang {thang} có 31 ngày"
elif thang == 4 or thang == 6 or thang == 9 or thang == 11:
    s = f"Thang {thang} có 30 ngày"
elif thang == 2:
    s = f"Thang {thang} có 28 hoặc 29 ngày"
else:
    s = f"Tháng {thang} không có trong năm"

print(s)
```

```python
thang = int(input("Nhập tháng: "))
nam = int(input("Nhập năm: "))
match thang:
    case 1 | 3 | 5 | 7 | 8 | 10 | 12:
        s = f"Thang {thang} có 31 ngày"
    case 4 | 6 | 9 | 11:
        s = f"Thang {thang} có 30 ngày"
    case 2:
        if nam % 4 == 0 and (nam % 100 != 0 or nam % 400 == 0):
            s = f"Thang {thang} có 29 ngày"
        else:
            s = f"Thang {thang} có 28 ngày"
    case _:
        s = f"Tháng {thang} không có trong năm"

print(s)
```
