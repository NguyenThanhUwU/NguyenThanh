W8A16: Diện tích các hình học
Yêu cầu:
Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

shape_type = input().strip()
params = list(map(float, input().split()))

if shape_type == "RECTANGLE":
    obj = Rectangle(params[0], params[1])
elif shape_type == "CIRCLE":
    obj = Circle(params[0])

print(f"{obj.area():.2f}")
```

Xây dựng lớp Car(brand, speed, seats) kế thừa từ Vehicle.
Viết phương thức info() cho mỗi lớp.

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def info(self):
        return f"{self.brand}-{self.speed}"

class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        super().__init__(brand, speed)
        self.seats = seats

    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"

obj_type = input().strip()
data = input().split()

if obj_type == "Vehicle":
    obj = Vehicle(data[0], data[1])
elif obj_type == "Car":
    obj = Car(data[0], data[1], data[2])

print(obj.info())
```

W8A18: Chu vi đa giác
Yêu cầu:
Xây dựng lớp Polygon với phương thức perimeter().
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c.

```python
class Polygon:
    def perimeter(self):
        pass

class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

a, b, c = map(int, input().split())
t = Triangle(a, b, c)
print(t.perimeter())
```

W8A20: Lãi suất tài khoản tiết kiệm
Yêu cầu:
Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance().
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức add_interest().

```python
class Account:
    def __init__(self, balance):
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        super().__init__(balance)
        self.rate = float(rate)

    def add_interest(self):
        self.balance += self.balance * self.rate

balance, rate = map(float, input().split())
acc = SavingAccount(balance, rate)
acc.add_interest()
print(f"{acc.get_balance():.2f}")
```

W8A21: Máy in số nhị phân

Xây dựng lớp BinaryPrinter kế thừa từ Printer, in biểu diễn nhị phân của số nguyên dương x.

```python
class Printer:
    def print_value(self, x):
        pass

class BinaryPrinter(Printer):
    def print_value(self, x):
        print(bin(x)[2:])

n = int(input())
p = BinaryPrinter()
p.print_value(n)
```

W8A22: Ma trận 2x2
Yêu cầu:
Xây dựng lớp Matrix2x2(a, b, c, d).
Viết phương thức trace() và determinant().

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d

    def trace(self):
        return self.a + self.d

    def determinant(self):
        return self.a * self.d - self.b * self.c

a, b, c, d = map(int, input().split())
m = Matrix2x2(a, b, c, d)
print(f"{m.trace()} {m.determinant()}")
```

W8A25: Tổng hai đa thức bậc nhất
Yêu cầu:
Xây dựng lớp PolynomialSum cho hai đa thức bậc nhất: P(x) = ax + b, Q(x) = cx + d.
Viết phương thức sum_str() trả về chuỗi biểu diễn đa thức tổng dạng kx + m.

```python
class PolynomialSum:
    def __init__(self, a, b, c, d):
        self.k = a + c
        self.m = b + d

    def sum_str(self):
        if self.m < 0:
            return f"{self.k}x - {-self.m}"
        return f"{self.k}x + {self.m}"

a, b, c, d = map(int, input().split())
p = PolynomialSum(a, b, c, d)
print(p.sum_str())
```

W8A27: Giá sản phẩm sau giảm giá
Yêu cầu:
Xây dựng lớp Product(name, price).
Xây dựng lớp DiscountProduct kế thừa từ Product, ghi đè final_price() để tính giá sau giảm.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = float(price)

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        super().__init__(name, price)
        self.percent = float(percent)

    def final_price(self):
        return self.price * (100 - self.percent) / 100

data = input().split()
name = data[0]
price = float(data[1])
percent = float(data[2])

p = DiscountProduct(name, price, percent)
print(f"{p.final_price():.2f}")
```

W8A28: Bộ đếm
Yêu cầu:
Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), value()

```python
class Counter:
    def __init__(self):
        self.count = 0

    def inc(self):
        self.count += 1

    def dec(self):
        self.count -= 1

    def value(self):
        return self.count

n = int(input())
ops = input().split()
c = Counter()

for op in ops:
    if op == "INC":
        c.inc()
    elif op == "DEC":
        c.dec()

print(c.value())
```

W8A29: Quy đổi thời gian
Yêu cầu:
Xây dựng lớp Time(h, m, s) với phương thức to_seconds() đổi thời gian sang tổng số giây.

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s

    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s

h, m, s = map(int, input().split())
t = Time(h, m, s)
print(t.to_seconds())
```

W8A30: Ngăn xếp (Stack)
Yêu cầu:
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), pop(), top().

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, x):
        self.items.append(x)

    def pop(self):
        if self.items:
            self.items.pop()

    def top(self):
        if self.items:
            return self.items[-1]

n = int(input())
st = Stack()

for _ in range(n):
    cmd = input().split()
    if cmd[0] == "PUSH":
        st.push(int(cmd[1]))
    elif cmd[0] == "POP":
        st.pop()
    elif cmd[0] == "TOP":
        print(st.top())
```
