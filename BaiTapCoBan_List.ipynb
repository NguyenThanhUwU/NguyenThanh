### Bài 1: Tìm số lớn nhất của 1 list

```python
def tim_max(lst):
    if not lst:
        return None
    max_val = lst[0]
    for num in lst[1:]:
        if num > max_val:
            max_val = num
    return max_val

my_list = [1, 5, 3, 9, 2]
print(f"List: {my_list}")
print(f"Max: {tim_max(my_list)}")
```

### Bài 2: Tìm số nhỏ nhất của 1 list

```python
def tim_min(lst):
    if not lst:
        return None
    min_val = lst[0]
    for num in lst[1:]:
        if num < min_val:
            min_val = num
    return min_val

my_list = [10, 5, 3, 9, 2]
print(f"List: {my_list}")
print(f"Min: {tim_min(my_list)}")
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó

```python
def tim_min_va_vi_tri(lst):
    if not lst:
        return None, None
    min_val = lst[0]
    min_idx = 0
    for i, num in enumerate(lst):
        if num < min_val:
            min_val = num
            min_idx = i
    return min_val, min_idx

my_list = [10, 0, 0, 12, 34, 5, 6, 7, 8, 89]
val, idx = tim_min_va_vi_tri(my_list)
print(f"List: {my_list}")
print(f"Min value: {val}, First index: {idx}")
```

### Bài 3: Tìm giá trị trung bình của 1 list

```python
def tinh_trung_binh(lst):
    if not lst:
        return 0.0
    return sum(lst) / len(lst)

my_list = [1, 2, 3, 4, 5]
print(f"List: {my_list}")
print(f"Average: {tinh_trung_binh(my_list)}")
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
def kiem_tra_doi_xung(lst):
    return lst == lst[::-1]

list_1 = [1, 2, 3, 2, 1]
list_2 = [1, 2, 3, 4, 5]
print(f"{list_1} doi xung? {kiem_tra_doi_xung(list_1)}")
print(f"{list_2} doi xung? {kiem_tra_doi_xung(list_2)}")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list

```python
def dua_khong_ve_dau(lst):
    ds_khong = [x for x in lst if x == 0]
    ds_khac_khong = [x for x in lst if x != 0]
    return ds_khong + ds_khac_khong

my_list = [3, 4, 6, 7, 1, 9, 0, 0, 23, 0, 0, 34, 67, 89]
print(f"Original: {my_list}")
print(f"Result: {dua_khong_ve_dau(my_list)}")
```

### Bài 6: Xóa tất cả các số giống nhau trong list (giữ nguyên vị trí)

```python
def xoa_trung_lap(lst):
    ket_qua = []
    seen = set()
    for so in lst:
        if so not in seen:
            ket_qua.append(so)
            seen.add(so)
    return ket_qua

my_list = [1, 2, 2, 3, 4, 4, 5, 1]
print(f"Original: {my_list}")
print(f"Unique: {xoa_trung_lap(my_list)}")
```

### Bài 7: Kiểm tra một list số nguyên là dãy tăng, dãy giảm hay vô danh

```python
def kiem_tra_day_so(lst):
    if len(lst) < 2:
        return "Danh sach qua ngan"
    is_tang = True
    is_giam = True
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]:
            is_tang = False
        if lst[i] <= lst[i+1]:
            is_giam = False
    if is_tang:
        return "Tang"
    elif is_giam:
        return "Giam"
    else:
        return "Vo danh"

print(f"[1, 3, 9] -> {kiem_tra_day_so([1, 3, 9])}")
print(f"[9, 3, 1] -> {kiem_tra_day_so([9, 3, 1])}")
print(f"[1, 2, 2, 3] -> {kiem_tra_day_so([1, 2, 2, 3])}")
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac không

```python
def kiem_tra_ziczac(lst):
    if len(lst) < 3:
        return "Khong phai ziczac"
    for i in range(len(lst) - 2):
        a, b, c = lst[i], lst[i+1], lst[i+2]
        is_peak = (b > a) and (b > c)
        is_valley = (b < a) and (b < c)
        if not (is_peak or is_valley):
            return "Khong phai ziczac"
    return "ziczac"

print(f"[1, 3, 2, 5] -> {kiem_tra_ziczac([1, 3, 2, 5])}")
print(f"[1, 2, 3, 4] -> {kiem_tra_ziczac([1, 2, 3, 4])}")
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k

```python
def tim_vi_tri_k(lst, k):
    vi_tri = [i for i, val in enumerate(lst) if val == k]
    return vi_tri if vi_tri else "-1"

lst = [1, 3, 4, 4, 3, 3, 4, 3, 6, 1]
print(f"List: {lst}")
print(f"k=3 positions: {tim_vi_tri_k(lst, 3)}")
print(f"k=10 positions: {tim_vi_tri_k(lst, 10)}")
```

### Bài 10: In ra kí tự đầu tiên và kí tự cuối cùng của tên

```python
def ky_tu_dau_cuoi(ten):
    ten = ten.strip()
    if not ten:
        return []
    return [ten[0], ten[-1]]

ten = "Nguyen Ai Quoc"
print(f"Ten: {ten} -> {ky_tu_dau_cuoi(ten)}")
```
