W2A6 [StringDataType] 
“Chuỗi” trong python là các kí tự được xâu lại với nhau như những hạt trên một chiếc dây chuyền. 
Chính vì thế “chuỗi” còn được gọi là “xâu”. Một chuỗi gồm các kí tự nằm trong dấu nháy đơn '' hoặc 
nháy đôi "”.• Mỗi chuỗi (str) là một chuỗi các kí tự trong bảng mã Unicode kể từ Python 3 trở đi 
(Unicode là một tiêu chuẩn mã hóa kí tự toàn cầu, cho phép biểu diễn hầu hết mọi kí tự trong tất cả 
các ngôn ngữ trên thế giới, bao gồm cả kí hiệu, biểu tượng, emoji,….). Bạn có thể dùng ord() và 
chr() để làm việc với mã Unicode.  
Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kí 
tự hoa tương ứng, in kí tự hoa này ra màn hình.

```python
ch = input("Nhập một ký tự thường (a-z): ")
print("Mã Unicode của", ch, "là:", ord(ch))
ch_upper = chr(ord(ch) - 32) 
print("Ký tự hoa tương ứng là:", ch_upper)
```

W2A8 [CelsiusTranfer] 
Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel 
Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy 
chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm 
tròn đến hai chữ số thập phân đằng sau dấu phẩy.

```python
a = float(input("Nhập độ C: "))
b = a * 9/5 + 32
print("Độ F tương ứng là:", b)
```

W2A9 [ShoppingOnline] 
Sơn Tùng mua một chiếc đồng hồ Movado trên trang mua sắm trực tuyến Amazon. Chiếc 
đồng hồ có giá x USD (x nhập từ bàn phím). Bởi Amazon không vận chuyển trực tiếp về 
Việt Nam, Đạt phải trả 10 USD cho phí vận chuyển từ một nước trung gian. Khi vận chuyển 
về Việt Nam, Đạt được thông báo sẽ phải trả thêm 30% thuế xuất nhập cảnh và 10% thuế 
VAT cho chiếc đồng hồ nói trên. Viết chương trình tính tổng số tiền mà Đạt phải trả để mua 
được chiếc đồng hồ trên (USD). Kết quả làm tròn đến hai chữ số thập phân. 

```python
a = float(input("Nhập số USD: "))
b = a + (a * 0.3) + (a * 0.1) + 10
print("Tổng số tiền mà đạt phải trả là:", b, "USD")
```

W2A10 [Friend Names] 
Viết chương trình nhập vào tên 3 người từ bàn phím.  
In ra màn hình tên theo thứ tự đảo ngược. 
VD: input ("Bob Alice Helen") thì output: ("Hi Helen, Alice and Bob.") 

```python
names = input("Nhập tên 3 người: ")
name_list = names.split()
name_list.reverse()
print(f"Hi {name_list[0]}, {name_list[1]} and {name_list[2]}.")
```

W2A11 Viết chương trình chuyển từ giờ và phút sang giây 
Đầu vào gồm hai giá trị chứa số giờ \h và số phút \m:  
Đầu ra: In ra màn hình một dòng duy nhất số giờ và phút được chuyển sang đơn vị giây. 

```python
a = float(input("Nhập giờ: "))
b = float(input("Nhập phút: "))
c = a * 3600 + b * 60
print("Tổng số giây là:", c, "giây")
```

W2A12 Cho một khối Rubik có độ dài cạnh là n, hãy trả về số lượng miếng dán riêng lẻ cần thiết để 
bao phủ toàn bộ khối lập phương. Ví dụ, khối rubic có độ dài cạnh là 1 cần 6 miếng dán 
riêng lẻ. (Lưu ý: một khối rubic có 6 mặt)

```python
n = float(input("Nhập độ dài cạnh của khối rubik: "))
s = 6 * n**2
print("Số lượng miếng dán cần thiết là:", s)
```

của phép tính tích giữa \a và \b.
