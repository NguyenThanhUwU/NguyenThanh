### **Mã bài:** W5A16
**Đề bài:** Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_armstrong(x):
    num_str = str(x)
    power = len(num_str)
    return x == sum(int(digit)**power for digit in num_str)

# Test
# print(is_armstrong(153))
```

### **Mã bài:** W5A17
**Đề bài:** Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
import math

def is_perfect_square(x):
    if x < 0:
        return False
    root = math.isqrt(x)
    return root * root == x

# Test
# print(is_perfect_square(16))
```

### **Mã bài:** W5A19
**Đề bài:** Viết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_magic_number(n):
    # Số kỳ diệu thường được định nghĩa là số có tổng các chữ số lặp lại bằng 1
    n = abs(n)
    while n > 9:
        n = sum(int(digit) for digit in str(n))
    return n == 1

# Test
# print(is_magic_number(19))
```

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'

# Test
# print(detect_number_type('123'))
# print(detect_number_type('12.3'))
# print(detect_number_type('abc'))
```

**Đề bài:** Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    # Sử dụng công thức diện tích tam giác = 0 để kiểm tra thẳng hàng
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)

# Test
# print(are_collinear(0, 0, 1, 1, 2, 2))
```

### **Mã bài:** W5A25
**Đề bài:** Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    # Sử dụng thuật toán Ray Casting cơ bản cho đa giác
    x, y = p
    n = len(quad)
    inside = False
    p1x, p1y = quad[0]
    for i in range(1, n + 1):
        p2x, p2y = quad[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xints = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xints:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside

# Test
# quad = [(0,0), (4,0), (4,3), (0,3)]
# p = (2,1)
# print(point_in_quadrilateral(quad, p))
```

### **Mã bài:** W5A27
**Đề bài:** Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_nonzero_digits(numbers):
    count = 0
    for num in numbers:
        count += sum(1 for digit in str(abs(num)) if digit != '0')
    return count

# Test
# nums = [102, 340, 50]
# print(count_nonzero_digits(nums))
```

### **Mã bài:** W5A28
**Đề bài:** Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def angle_ABC(A, B, C):
    # Tính độ dài 3 cạnh
    c = math.dist(A, B)  # AB
    a = math.dist(B, C)  # BC
    b = math.dist(A, C)  # AC
    
    # Áp dụng định lý cosin
    if a == 0 or c == 0:
        return 0.0
    cos_B = (a**2 + c**2 - b**2) / (2 * a * c)
    
    # Đảm bảo cos_B nằm trong [-1, 1] do sai số dấu phẩy động
    cos_B = max(-1.0, min(1.0, cos_B))
    angle_rad = math.acos(cos_B)
    
    return math.degrees(angle_rad)

# Test
# A, B, C = (0, 1), (0, 0), (1, 0)
# print(angle_ABC(A, B, C))
```

### **Mã bài:** W5A33
**Đề bài:** Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Đặt các hằng số quy tắc mặc định vì đề chưa cung cấp chi tiết số liệu
    base_fare = 10000  # Giá mở cửa
    price_per_km = 15000
    wait_fee_per_min = 1000
    
    total = base_fare + (km * price_per_km) + (wait_minutes * wait_fee_per_min)
    
    # Phụ phí ban đêm (từ 22h đến 5h sáng)
    if start_hour >= 22 or start_hour <= 5:
        total *= 1.2  # Tăng 20%
        
    # Có thể thêm các quy tắc giảm giá khác tùy thuộc yêu cầu thực tế
    return total

# Test
# print(smart_taxi_fare('4_cho', 23, 10, 5))
```

### **Mã bài:** W5A34
**Đề bài:** Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n <= 0 trả về 0.

```python
def has_unique_digits(x):
    s = str(x)
    return len(s) == len(set(s))

def sum_1_to_n(n):
    if n <= 0:
        return 0
    return sum(x for x in range(1, n + 1) if has_unique_digits(x))

# Test
# print(sum_1_to_n(15))
```
