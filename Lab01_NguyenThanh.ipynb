### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python

n = int(input())
def count_k(n):
    if n <= 1:
        return 0
    
    return (n - 1) - ((n - 1) // 2)
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def shift_char(x, n):
    if not x.isalpha():
        return x
    
    start = ord('A') if x.isupper() else ord('a')
    new_pos = (ord(x) - start + n) % 26
    return chr(start + new_pos)

if __name__ == "__main__":
    char_input = input("Nhập ký tự: ")
    shift_input = int(input("Nhập số vị trí dịch chuyển (n): "))

    if len(char_input) > 0:
        result = shift_char(char_input[0], shift_input)
        print(f"Kết quả: {result}")
```
