Mã bài W6A16: Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải.

```python
def W6A16():
    n, m = map(int, input().split())
    matrix = []
    for _ in range(n):
        matrix.append(list(map(int, input().split())))
    for row in matrix:
        for val in row:
            print(f"{val:>4}", end="")
        print()
```

Mã bài W6A17: Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng.

```python
def W6A17():
    n = int(input())
    matrix = []
    for _ in range(n):
        matrix.append(list(map(int, input().split())))
    main_diag = [matrix[i][i] for i in range(n)]
    anti_diag = [matrix[i][n - 1 - i] for i in range(n)]
    print(*main_diag)
    print(*anti_diag)
```

Mã bài W6A18: Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k).

```python
def W6A18():
    first_line = list(map(int, input().split()))
    n, m, k = first_line[0], first_line[1], first_line[2]
    matrix = []
    for _ in range(n):
        matrix.append(list(map(int, input().split())))
    col_sum = sum(matrix[i][k] for i in range(n))
    row_sum = sum(matrix[k])
    print(col_sum, row_sum)
```

Mã bài W6A20-1: Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1.

```python
def W6A20_1(lst, k):
    try:
        return lst.index(k)
    except ValueError:
        return -1
```

Mã bài W6A21: Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
def W6A21():
    words = input().split()
    pos_dict = {}
    for i, word in enumerate(words):
        if word in pos_dict:
            pos_dict[word] = pos_dict[word] + (i,)
        else:
            pos_dict[word] = (i,)
    return pos_dict
```

Mã bài W6A24: Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 02324000900023249.

```python
def W6A24(lst):
    zeros = [val for val in lst if val == 0]
    others = [val for val in lst if val != 0]
    return zeros + others
```

Mã bài W6A25: Cho dãy chỉ gồm N giá trị từ 0-9 . Kiểm tra xem giá trị nào không có trong mảng

```python
def W6A25(lst):
    all_digits = set(range(10))
    present_digits = set(lst)
    missing = sorted(list(all_digits - present_digits))
    return missing
```

Mã bài W6A26: Tính tông các phân tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def W6A26(lst):
    total = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i - 1] and lst[i] > lst[i + 1]:
            total += lst[i]
    return total
```

Mã bài W6A27: Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1.

```python
def W6A27(lst, x):
    indices = [i for i, val in enumerate(lst) if val < x]
    if not indices:
        print(-1)
    else:
        print(*indices)
```

Mã bài W6A28: Cho một dãy các số nguyên, kiểm tra xem nó là dây tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
def W6A28(lst):
    if len(lst) <= 1:
        return "vô danh"
    is_increasing = all(lst[i] < lst[i+1] for i in range(len(lst)-1))
    is_decreasing = all(lst[i] > lst[i+1] for i in range(len(lst)-1))
    
    if is_increasing:
        return "tăng thực sự"
    elif is_decreasing:
        return "giảm thực sự"
    else:
        return "vô danh"
```
