Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
import math

a = float(input())
b = float(input())
c = float(input())

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Deu")
    elif a == b or b == c or a == c:
        print("Can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Vuong")
    else:
        print("Thuong")
    
    cv = a + b + c
    p = cv / 2
    dt = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    print(dt)
    print(cv)
else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math

a = float(input())
b = float(input())
c = float(input())

if a == 0:
    if b == 0:
        if c == 0:
            print("Vo so nghiem")
        else:
            print("Vo nghiem")
    else:
        print(-c / b)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Vo nghiem")
    elif delta == 0:
        print(-b / (2*a))
    else:
        print((-b + math.sqrt(delta)) / (2*a))
        print((-b - math.sqrt(delta)) / (2*a))
```

Bài 8: Giải hệ phương trình tuyến tính: 
ax  by  m 
cx  dy  n

```python
a = float(input())
b = float(input())
m = float(input())
c = float(input())
d = float(input())
n = float(input())

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D == 0:
    if Dx == 0 and Dy == 0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
else:
    print(Dx / D)
    print(Dy / D)
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không? 

```python
import math

x0 = float(input())
y0 = float(input())
r = float(input())
xa = float(input())
ya = float(input())

d = math.sqrt((xa - x0)**2 + (ya - y0)**2)

if math.isclose(d, r):
    print("Thuoc duong tron")
else:
    print("Khong thuoc duong tron")
```
